```python
url = "https://coinmarketcap.com/currencies/electroneum/historical-data/?start=20130428&end=20190512"
df = pd.read_html(url)
proc = df[0]
print(proc)
```

```python
proc["Date"] = pd.DatetimeIndex(data=proc["Date"])
proc.dtypes
```

```python
plt.figure(figsize=(10, 7))
plt.plot(proc["Date"], proc["Close**"])
plt.show()
```
